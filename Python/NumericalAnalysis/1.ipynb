```python
def find_max(count, data):
    max = data[0]
    i = 0
    for i in range(count):
        if(data[i] > max):
            max = data[i]
    return max

def main():
        user_input = input("请输入一组数字（空格分隔）: ")
        array = list(map(int, user_input.split()))
        num = int(input(f"第几个(1-{len(array)}):"))
        result = find_max(num, array)
        print(f"结果是: {result}")
if __name__ == "__main__": 
    main()
```

```python
def fibs(n):
    if n == 1:
        return [1]
    if n == 2:
        return [1,1]
    fibs = [1,1]
    for i in range(2,n):
        fibs.append(fibs[i-1]+fibs[i-2])
    return fibs

print(fibs(10))
```

```python
def roots(a,b,c):
    delta = sqrt(b**2-4*a*c)
    x1 = (-b + delta)/(2*a)
    x2 = (-b - delta)/(2*a)
    return x1, x2
a ,b, c = 1, -2, -3
x1, x2 = roots(a,b,c)
print(f'{a}x^2+{b}x+{c}=0的根为{x1}{x2}')
```
