##### 2. 利用numpy库求解下面的矩阵方程：

$$
\begin{pmatrix}
1 & 4  \\
-1 & 2 \\
\end{pmatrix}X
\begin{pmatrix} 
2 & 0 \\
-1 & 1 \\ 
\end{pmatrix} 
= \begin{pmatrix} 
3 & 1\\ 
0 & -1 \end{pmatrix}.
$$

**提示**：设上面的矩阵方程为$AXB=C$，则$X=A^{-1}CB^{-1}$.

```python
# code here
A=np.array([[1,4],[-1,2]])
B=np.array([[2,0],[-1,1]])
C=np.array([[3,1],[0,-1]])
An=np.linalg.inv(A)
Bn=np.linalg.inv(B)
X=An@C@Bn
print("矩阵X=\n",X)
```

##### 5. 编程判断$n=199933$是否为素数。

**提示**：设整数$a\in [2, \sqrt{n}]$，如果存在$a$使得$a|n$，则$n$不是素数。反之，则$n$是素数。

```python
# code here
import math
def solve(n):
    for a in range(2, int(math.sqrt(n)) + 1):
        if (n % a) == 0:
            return False
    return True
n = 199933
if solve(n):
    print(f"n={n} 是素数")
else:
    print(f"n={n} 不是素数")
```

##### 6. **倒数递归**：根据下面的迭代序列，
$$
x_n = \frac{1}{x_{n-1}} + \frac{1}{x_{n-2}}, n=2,3,\cdots
$$
编写`recip_iteration(x0, x1, epsilon)`函数，其中$x_0$和$x_1$为迭代初始值，当$|x_{n}-x_{n-1}|<\text{epsilon}$时停止迭代，返回最后的迭代值$x_n$和迭代进行的总次数。

（1）取$x_0=1$，$x_1=2$，$\text{epsilon}=0.5e-8$，输出迭代值。

（2）测试其它初始值$x_0$和$x_1$，猜测得出你发现的结论。


```python
# code here
def recip_iteration(x0, x1, epsilon):
    cmp = 0
    while 1:
        xn = 1 / x1 + 1 / x0
        cmp += 1
        if abs(xn - x1) < epsilon:
            return xn, cmp
        x0, x1 = x1, xn
x0, x1, epsilon = 1, 2, 0.5e-8
result, times = recip_iteration(x0, x1, epsilon)
print(f"迭代值: {result:.10f}, 迭代数: {times}")
x0, x1 = 14, 122
result, times = recip_iteration(x0, x1, epsilon)
print(f"其他初始迭代值: {result:.10f}, 迭代数: {times}")
```

**结果分析**：结果都趋于$\sqrt{2}$

##### 8. 平面图形变换：旋转、平移和放缩

旋转是一种常见的图形变换，它将图形绕着原点进行旋转。在二维$xOy$坐标系下，点$P(x,y)$逆时针旋转$\alpha$后为点$P'(x',y')$，利用复数知识或平面几何知识可得

$$
\begin{cases}
x' = x\cos \alpha -y\sin \alpha,    \\
y' = x\sin \alpha +y\cos \alpha
\end{cases}
$$

写成矩阵的形式，有
$$
\begin{pmatrix}
x' \\
y'
\end{pmatrix} = 
\begin{pmatrix}
\cos \alpha & -\sin\alpha \\
\sin\alpha & \cos\alpha 
\end{pmatrix} 
\begin{pmatrix}
x 
\\y
\end{pmatrix}\\

\triangleq R({\alpha})\begin{pmatrix}
x 
\\y
\end{pmatrix}
$$
其中，$R({\alpha}) =\begin{pmatrix}
\cos \alpha & -\sin\alpha \\
\sin\alpha & \cos\alpha 
\end{pmatrix}  $称为旋转矩阵。
